### configuring the driver to scrap the table from the page for each of the three seasons and clicking the NEXT buttom if there is another page

```python

# Configura il driver con timeout più alto
driver.set_page_load_timeout(45)  # Aspetta fino a 25 secondi per il caricamento della pagina

all_data = []

try:
    while True:
        # Aspetta che la tabella sia caricata
        WebDriverWait(driver, 15).until(EC.presence_of_element_located((By.TAG_NAME, 'table')))
        
        # Trova la tabella e le righe
        table = driver.find_element(By.TAG_NAME, 'table')
        rows = table.find_elements(By.TAG_NAME, 'tr')
        
        for row in rows:
            cells = row.find_elements(By.TAG_NAME, 'td')
            cells_text = [cell.text.strip() for cell in cells]
            if cells_text:
                all_data.append(cells_text)

        # Scorri fino in fondo per assicurarti che il pulsante "Next" sia visibile
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)

        # **🔹 Seleziona il pulsante "Next" esattamente evitando "Previous"**
        try:
            pagination = driver.find_element(By.CLASS_NAME, "pagination")
            next_buttons = pagination.find_elements(By.CLASS_NAME, "button")
            
            # Trova l'elemento "Next" tra i pulsanti
            next_button = None
            for btn in next_buttons:
                if "Next" in btn.text:
                    next_button = btn
                    break
            
            if next_button is None:
                print("Nessun pulsante 'Next' trovato. Fine delle pagine.")
                break  
        
        except:
            print(" Nessun pulsante 'Next' trovato. Fine delle pagine.")
            break  

        # Scroll fino al pulsante per renderlo visibile
        driver.execute_script("arguments[0].scrollIntoView();", next_button)
        time.sleep(1)

        # Usa JavaScript per cliccare se il click classico fallisce
        driver.execute_script("arguments[0].click();", next_button)

        # Aspetta che la nuova pagina carichi i dati
        WebDriverWait(driver, 25).until(EC.staleness_of(table))

    # Converti i dati in DataFrame
    df = pd.DataFrame(all_data)

    # Salva in un file CSV
    df.to_csv('all_players_2023_24.csv', index=False, encoding='utf-8')
    print("Dati di tutte le pagine salvati in all_players_2023.csv")

finally:
    driver.quit()
```

### Separating Player Names and Roles in the Dataset

```python
# Lista dei ruoli conosciuti
roles_list = ["CB", "LB", "RB", "ST", "LW", "RW", "CM", "CDM", "CAM", "CF", "RM", "LM", "LWB", "RWB","GK"]

# Funzione per separare nome e ruoli
def split_name_and_roles_safe(name):
    # Dividi la stringa in parole
    parts = name.split()
    # Inizializza liste per nome e ruoli
    roles = []
    name_parts = []
    # Itera sulle parole dalla fine all'inizio
    for word in reversed(parts):
        if word in roles_list:  # Se la parola è un ruolo, aggiungila ai ruoli
            roles.insert(0, word)
        else:  # Altrimenti, considera la parola parte del nome
            name_parts.insert(0, word)
    # Ricostruisci nome e ruoli
    name_only = " ".join(name_parts)
    roles_only = ", ".join(roles)
    return name_only, roles_only
    

# Leggi il file CSV (sostituisci con il tuo file)
file_path = "players_all_3_seasons.csv"  # Modifica con il percorso del tuo file
data = pd.read_csv(file_path)

# Applica la funzione alla colonna 'Name'
data[['Name', 'Position']] = data['Name'].apply(lambda x: pd.Series(split_name_and_roles_safe(x)))

# Riorganizza le colonne per posizionare "Roles" come seconda
columns_order = ['Name', 'Position'] + [col for col in data.columns if col not in ['Name', 'Position']]
data = data[columns_order]

# Salva il risultato in un nuovo file
output_file = "statistics_players.csv"
data.to_csv(output_file, index=False)

print(f"I dati separati sono stati salvati in: {output_file}")
```

### splitting the team & contract column and converting the value and wage of the players using the convert currency and converted the height in cm and weight in kg

```python
players_all_3_seasons = pd.read_csv("statistics_players.csv")

players_all_3_seasons[['Team', 'Contract']] = players_all_3_seasons['Team & Contract'].str.split('\n', expand=True)


# Function to convert values
def convert_currency(value):
    if 'M' in value:
        return float(value.replace('€', '').replace('M', '')) * 1000000
    elif 'K' in value:
        return float(value.replace('€', '').replace('K', '')) * 1000
    else:
        return float(value.replace('€', ''))
    return value

# Apply the function to the column
players_all_3_seasons['Converted Value'] = players_all_3_seasons['value'].apply(convert_currency)
players_all_3_seasons['Converted Wage'] = players_all_3_seasons['Wage'].apply(convert_currency)

players_all_3_seasons['Height (cm)'] = players_all_3_seasons['Height'].str.extract(r'(\d+)cm').astype(int)
players_all_3_seasons['Weight (kg)'] = players_all_3_seasons['Weight'].str.extract(r'(\d+)kg').astype(int)


dic_positions = {
    'GK': 'Goalkeeper', 
    'LB': 'Left-Back', 
    'LWB': 'Left-Back', 
    'LCB': 'Centre-Back', 
    'CB': 'Centre-Back',
    'RCB': 'Centre-Back', 
    'RB': 'Right-Back', 
    'RWB': 'Right-Back', 
    'LDM': 'Defensive Midfield',
    'CDM': 'Defensive Midfield', 
    'RDM': 'Defensive Midfield', 
    'LM': 'Left Midfield', 
    'LCM': 'Central Midfield',
    'CM': 'Central Midfield', 
    'RCM': 'Central Midfield', 
    'RM': 'Right Midfield', 
    'LAM': 'Attacking Midfield',
    'CAM': 'Attacking Midfield', 
    'RAM': 'Attacking Midfield', 
    'LW': 'Left Winger', 
    'LF': 'Forward',
    'CF': 'Forward', 
    'RF': 'Forward', 
    'LS': 'Forward', 
    'ST': 'Forward',
    'RW': 'Right Winger'
}

# Display the updated DataFrame
display(players_all_3_seasons)
```

### Standardizing and Replacing Player Positions in the Dataset based on the dictionary

```python
# Funzione per sostituire i ruoli nella colonna Position
def replace_positions(positions):
    roles = positions.split(', ')  # Dividi i ruoli separati da una virgola
    updated_roles = [dic_positions.get(role, role) for role in roles]  # Sostituisci i ruoli
    updated_roles = list(set(updated_roles))  # Rimuovi i duplicati
    updated_roles.sort()  # Ordina i ruoli in ordine alfabetico (opzionale)
    return ', '.join(updated_roles)  # Ritorna i ruoli sostituiti come una stringa separata da virgole

# Applicare la funzione alla colonna 'Position'

# Sostituire i valori nella colonna "Position" usando il dizionario
players_all_3_seasons['Position'] = players_all_3_seasons['Position'].apply(replace_positions)
```

### Normalizing Player Names and Teams: Lowercasing and Removing Accents

```python
import re
from unidecode import unidecode

# Function to abbreviate name: first initial + last part
def abbreviate_name(name):
    if isinstance(name, str):
        name = unidecode(name.lower())
        name = re.sub(r"[^a-zA-Z.\s]", "", name)  # tieni anche i punti
        parts = name.split()
        if len(parts) >= 2:
            # se il primo elemento è già un'iniziale (tipo "r.") la lasciamo così
            first = parts[0] if parts[0].endswith('.') else parts[0][0] + '.'
            last = parts[-1]
            return f"{first}{last}"
        return name
    return name

# Apply to player names
players_all_3_seasons['Name'] = players_all_3_seasons['Name'].apply(abbreviate_name)

# Normalize team names
players_all_3_seasons['Team'] = players_all_3_seasons['Team'].apply(
    lambda x: unidecode(x.lower()) if isinstance(x, str) else x
)
```

### creating the ID by removing the point in the name of the players and create the ID concatanating the name,the team and the season using the underscore

```python
# Funzione per creare l'ID
def create_player_id(row):
    # Rimuoviamo il punto nel nome
    name_parts = row['Name'].replace('.', ' ').split()

    # Verifica che il nome abbia almeno due parti (nome e cognome)
    if len(name_parts) >= 2:
        first_name_initial = name_parts[0][0].lower()  # Prima lettera del nome
        last_name = name_parts[1].lower()  # Cognome
    else:
        # Se non c'è un cognome, usiamo l'intero nome (senza punto) per la parte finale
        first_name_initial = name_parts[0][0].lower()
        last_name = name_parts[0][1:].lower()  # Aggiunge il resto del nome

    # Creiamo l'ID come lettera_cognome_squadra_stagione
    player_id = f"{first_name_initial}_{last_name}_{row['Team'].replace('.', ' ').replace(' ', '_').lower()}_{row['Season'].replace('/', ' ').replace(' ', '_').lower()}"
    
    return player_id

# Applichiamo la funzione per creare la colonna 'ID'
players_all_3_seasons['ID'] = players_all_3_seasons.apply(create_player_id, axis=1)

# Mostriamo il risultato
print(players_all_3_seasons)
```

```python
def process_positions(file_path, output_path):
    # Carica il file CSV
    df = pd.read_csv(file_path)
    
    # Verifica i nomi delle colonne
    print("Colonne del dataset:", df.columns)
    
    # Assicurati che la colonna "Position" esista
    if "Position" not in df.columns:
        raise ValueError("La colonna 'Position' non esiste nel dataset.")
    
    # Funzione per elaborare le posizioni
    def split_positions(position):
        # Assicurati che la posizione sia trattata come stringa
        position = str(position).strip()
        
        # Se la posizione è vuota, ritorna valori vuoti
        if not position:
            return pd.Series(["", "", 0])
        
        # Split delle posizioni, rimuovendo gli spazi extra
        positions = [pos.strip() for pos in position.split(",")]
        
        # La prima posizione è la posizione principale
        main_position = positions[0] if positions else ""
        
        # Le posizioni alternative sono tutte le posizioni eccetto la principale
        alternative_positions = ", ".join(positions[1:]) if len(positions) > 1 else ""
        
        # Conta le posizioni alternative
        alternative_count = len(positions) - 1 if len(positions) > 1 else 0
        
        return pd.Series([main_position, alternative_positions, alternative_count])
    
    # Applica la funzione a tutte le righe
    df[["Main Position", "Alternative Positions", "Alternative Count"]] = df["Position"].apply(split_positions)
    
    # Drop delle colonne "Position" e "Weight" se esistono
    df.drop(columns=[col for col in ["Position", "Weight","Team & Contract","Height"] if col in df.columns], inplace=True)
    
    # Salva il nuovo dataset
    df.to_csv(output_path, index=False)
    print(f"File salvato con successo: {output_path}")

# Esempio di utilizzo
input_file = "players_all_3_seasons.csv"  # Sostituisci con il percorso corretto
output_file = "filtered_with_seperation.csv"
process_positions(input_file, output_file)
```
